```python
df = pd.read_excel('/content/final_books.xlsx')
```

```python
# Normalize Data

df_normalized = []

normalizer = Normalizer()

for row in df['content']:
  if type(row) != str and pd.isna(row):
    row = ""
  df_normalized.append(normalizer.normalize(row))
```

```python
# Tokenize Data

df_tokenized = []

for row in df_normalized:
  df_tokenized.append(sent_tokenize(row))
```

```python
# Show Result

print("Content of First Row - Not Normalized - Not Tokenized")
print(df['content'][0])

print("\nContent of First Row - Normalized - Not Tokenized")
print(df_normalized[0:1])

print("\nContent of First Row - Normalized - Tokenized")
print(df_tokenized[0])
```

```python
!wget http://www.cs.cmu.edu/~ark/personas/data/MovieSummaries.tar.gz
!tar -xf MovieSummaries.tar.gz
clear_output()
```

```python
df = pd.read_csv("/content/MovieSummaries/plot_summaries.txt", delimiter = "\t",names=["id","text"])
```

```python
# Normalize Data

df_normalized = []

for row in df['text']:
  if type(row) != str and pd.isna(row):
    row = ""
  df_normalized.append(re.sub(r'[^\w\s]','',row))
```

```python
# Tokenize Data

df_tokenized = []

for row in df_normalized:
  df_tokenized.append(sent_tokenize(row))
```

```python
# Show Result

print("Content of First Row - Not Normalized - Not Tokenized")
print(df['text'][0])

print("\nContent of First Row - Normalized - Not Tokenized")
print(df_normalized[0:1])

print("\nContent of First Row - Normalized - Tokenized")
print(df_tokenized[0])
```
